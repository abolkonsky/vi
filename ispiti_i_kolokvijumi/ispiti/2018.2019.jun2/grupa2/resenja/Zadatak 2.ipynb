# Zadatak 2 - Minimax

Dva igrača igraju igru. Na početku igre svaki igrač dobija po 4 žetona i unosi se niz 8 celih brojeva (mogu biti i negativni). U svakom potezu igrač stavlja žeton na jedno polje niza. Kada oba igrača odigraju sve poteze sabiraju se polja na kojima se nalaze žetoni i pobedio je onaj igrač koji ima najviše u zbiru. 

Napisati minimax implementaciju ove igre. Na početku igre igra čovek.

```python
import copy

class Game:
    def __init__(self, niz):
        self.niz = niz
        self.suma_coveka = 0
        self.suma_masine = 0
        
        # Prvi igrac je na potezu (covek)
        self.player_turn = 1
        
        
    def ispisi_niz(self):
        print(self.niz)
        print('Suma coveka: {}'.format(self.suma_coveka))
        print('Suma masine: {}'.format(self.suma_masine))
        
    def valid(self, indeks):
        if indeks < 0 or indeks > 7 or self.niz[indeks] == '_':
            return False
        else:
            return True
     
    #ispitivanje da li se doslo do kraja igre
    def kraj(self):
        flag = 0
        for i in range(0, 8):
            if self.niz[i] != '_':
                flag = 1
                break
                
        if flag:
            return False
        else:
            return True
    
    # igrac Max je masina
    def Max(self):
        maxv = -2
        indeks = None
        
        if self.kraj():
            if self.suma_coveka > self.suma_masine:
                return (-1, _)
            elif self.suma_coveka == self.suma_masine:
                return (0, _)
            else:
                return (1, _)
        
        
        for i in range(0, 8):
            if self.niz[i] != '_':
                broj = self.niz[i]
                self.niz[i] = '_'
                self.suma_masine += broj
                m = self.Min()
                if m > maxv:
                    maxv = m
                    indeks = i
                self.niz[i] = broj
                self.suma_masine -= broj
                    
        return (maxv, indeks)
    
    
    def Min(self):
        minv = 2
        
        if self.kraj():
            if self.suma_coveka > self.suma_masine:
                return -1
            elif self.suma_coveka == self.suma_masine:
                return 0
            else:
                return 1  
        
        for i in range(0, 8):
            if self.niz[i] != '_':
                broj = self.niz[i]
                self.niz[i] = '_'
                self.suma_coveka += broj
                (m, indeks) = self.Max()
                if m < minv:
                    minv = m
                self.niz[i] = broj
                self.suma_coveka -= broj
                    
        return minv
    
        
    def play(self):
        
        while True:
            self.ispisi_niz()

            if self.kraj():
                if self.suma_coveka > self.suma_masine:
                    print('Pobedio je covek!')
                    return (-1, _)
                elif self.suma_coveka == self.suma_masine:
                    print('Nereseno!')
                    return (0, _)
                else:
                    print('Pobedila je masina')
                    return (1, _)  

            # ako je na potezu igrac 1, odnosno covek
            if self.player_turn == 1:

                while True:
                    indeks = int(input('Unesi indeks u nizu: '))
                    indeks -= 1

                    if self.valid(indeks):
                        self.suma_coveka += self.niz[indeks]
                        self.niz[indeks] = '_'
                        self.player_turn = 0
                        break;
                    else:
                        print('Potez nije validan')
             
            # ako je na potezu igrac O, odnosno masina
            else:
                (m, indeks) = self.Max()
                self.suma_masine += self.niz[indeks]
                self.niz[indeks] = '_'
                self.player_turn = 1
```
