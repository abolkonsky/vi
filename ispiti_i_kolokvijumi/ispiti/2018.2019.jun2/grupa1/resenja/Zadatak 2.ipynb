# Zadatak 2 - Minimax

_Nim_ je igra kinsekog porekla u kojoj najčešće učestvuju dva igrača i imaju tri hipa sa različitim
brojem kuglica. Igrači igraju naizmenično i u jednom potezu mogu skinuti tačno sa jednog hipa bilo koji broj kuglica. Cilj je biti poslednji igrač koji može da skine kuglice.

Napisati minimax implementaciju igre _nim_.

Igra treba da izgleda na sledeći način:
* Na početku postavlja se pitanje koliko na svakom hipu ima kuglica (za svaki hip po jedan broj).
* Nakon toga igru započinje igrač (čovek) i naizmenično igraju čovek i mašina.
* Nakon svakog poteza ispisuje se koliko je preostalo kuglica na svakom hipu.
* Na kraju se ispisuje ko je pobednik.

```python
import copy

class Game:
    def __init__(self, num1, num2, num3):
        
        self.hips = [num1, num2, num3]
        
        # Prvi igrac je na potezu
        # 1 oznacava da je na potezu covek
        self.player_turn = 1
        
    def ispisi_hipove(self):
        print('hip 1: {}'.format(self.hips[0]))
        print('hip 2: {}'.format(self.hips[1]))
        print('hip 3: {}'.format(self.hips[2]))
        
    def valid(self, hip, num):
        if hip < 0 or hip > 2 or self.hips[hip] < num:
            return False
        else:
            return True
     
    #ispitivanje da li se doslo do kraja igre
    def kraj(self):
        if self.hips[0] == 0 and self.hips[1] == 0 and self.hips[2] == 0:
            return True
        else:
            return False
    
    # igrac Max je racunar
    def Max(self):
        maxv = -2
        hip = None
        num = None
        
        if self.kraj():
            return (-1, _, _)
    
        
        for i in range(0, 3):
            for j in range(1, self.hips[i]+1):
                self.hips[i] -= j
                m = self.Min()
                if m > maxv:
                    maxv = m
                    hip = i
                    num = j
                self.hips[i] += j
                
        return (maxv, hip, num)
    
    
    def Min(self):
        minv = 2
        
        if self.kraj():
            return 1
        
        for i in range(0, 3):
            for j in range(1, self.hips[i]+1):
                self.hips[i] -= j
                (m, max_hip, max_num) = self.Max()
                if m < minv:
                    minv = m
                # Vracamo polje ponovo na prazno
                self.hips[i] += j     
                    
        return minv
    
        
    def play(self):
        
        while True:
            self.ispisi_hipove()

            if self.kraj():
                if self.player_turn == 1:
                    print('Pobedio je racunar!')
                    return
                else:
                    print('Pobedio je covek!')
                    return

            # ako je na potezu igrac X, odnosno covek
            if self.player_turn == 1:

                while True:
                    hip = int(input('Unesi broj hipa: '))
                    num = int(input('Unesi broj kuglica koji se skida sa hipa: '))
                    hip -= 1
                    
                    if self.valid(hip, num):
                        self.hips[hip] -= num
                        self.player_turn = 0
                        break;
                    else:
                        print('Potez nije validan')
             
            # ako je na potezu igrac O, odnosno masina
            else:
                (m, hip, num) = self.Max()
                print('Racunar skida sa {}. hipa skida {} kuglice'.format(hip+1, num))
                self.hips[hip] -= num
                self.player_turn = 1
```

```python
def Num():
    broj1 = int(input('Unesi broj kuglica na hipu1: '))
    broj2 = int(input('Unesi broj kuglica na hipu2: '))
    broj3 = int(input('Unesi broj kuglica na hipu3: '))
    g = Game(broj1, broj2, broj3)
    g.play()
```
